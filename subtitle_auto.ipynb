```python
import cv2
import numpy as np
from PIL import Image, ImageDraw, ImageFont
import os
from tqdm.notebook import tqdm

# --- 色計算用のヘルパー関数 ---
def get_luminance(bgr):
    rgb = np.array(bgr[::-1]) / 255.0
    a = [val / 12.92 if val <= 0.03928 else ((val + 0.055) / 1.055) ** 2.4 for val in rgb]
    return 0.2126 * a[0] + 0.7152 * a[1] + 0.0722 * a[2]

def get_contrast_ratio(lum1, lum2):
    lighter = max(lum1, lum2)
    darker = min(lum1, lum2)
    return (lighter + 0.05) / (darker + 0.05)

# --- テキストを自動で改行する関数 ---
def wrap_text(text, font, max_width):
    """長いテキストをmax_widthに合わせて改行し、行のリストを返す"""
    lines = []
    # 日本語は単語の区切りがないため、1文字ずつチェックする
    if not text:
        return []
        
    current_line = ""
    for char in text:
        # 1文字追加してみて、幅を超えるかテスト
        test_line = current_line + char
        # getbboxを使って正確な幅を取得
        text_bbox = font.getbbox(test_line)
        w = text_bbox[2] - text_bbox[0]

        if w <= max_width:
            current_line = test_line
        else:
            # 幅を超えたら、前の行をリストに追加して新しい行を開始
            lines.append(current_line)
            current_line = char
            
    lines.append(current_line) # 最後の行を追加
    return lines

# --- 準備 ---
%store -r file_name
%store -r result

# ★変更点：候補色を「明るい」「暗い」の2グループに分ける
LIGHT_COLORS = {
    'オフホワイト': (245, 245, 245),
    'ペールイエロー': (170, 232, 238),
    'ライトシアン': (238, 238, 175),
}

DARK_COLORS = {
    'チャコール': (79, 69, 54),
    'ダークブルー': (139, 61, 72),
    'ワインレッド': (42, 42, 128), # 中間の暗さの色を追加
}
# 各グループの輝度を事前に計算
LIGHT_LUMINANCES = {name: get_luminance(bgr) for name, bgr in LIGHT_COLORS.items()}
DARK_LUMINANCES = {name: get_luminance(bgr) for name, bgr in DARK_COLORS.items()}

# ファイルパス定義...
input_video_path = os.path.join('project/input_video', file_name)
video_only_path = os.path.join('project/tmp', f"video_only_{file_name}")
font_path = '/usr/share/fonts/opentype/noto/NotoSansCJK-Regular.ttc'

input_video = cv2.VideoCapture(input_video_path)
if not input_video.isOpened():
    print("エラー: 動画ファイルを開けませんでした。")
else:
    fps = input_video.get(cv2.CAP_PROP_FPS)
    total_frames = int(input_video.get(cv2.CAP_PROP_FRAME_COUNT))
    width = int(input_video.get(cv2.CAP_PROP_FRAME_WIDTH))
    height = int(input_video.get(cv2.CAP_PROP_FRAME_HEIGHT))

    fourcc = cv2.VideoWriter_fourcc(*'mp4v')
    output_video = cv2.VideoWriter(video_only_path, fourcc, fps, (width, height))
    font_size = 36
    font = ImageFont.truetype(font_path, font_size)

    subtitle_segments = result['segments']

    print("字幕の焼き付け処理を開始します...")
    for frame_num in tqdm(range(total_frames), desc="映像処理中"):
        ret, frame = input_video.read()
        if not ret: break

        current_time = frame_num / fps
        active_subtitle = None
        for segment in subtitle_segments:
            if segment['start'] <= current_time < segment['end']:
                active_subtitle = segment['text']
                break

        if active_subtitle:
            # --- wrap_text関数を呼び出し、複数行に対応 ---
            # 左右に少し余白(padding)を持たせる
            margin = 40 
            wrapped_lines = wrap_text(active_subtitle, font, width - margin)
            
            # 複数行のテキスト全体の高さを計算
            sample_bbox = font.getbbox("あ") # 1文字分の高さを取得
            line_height = sample_bbox[3] - sample_bbox[1]
            total_text_height = len(wrapped_lines) * line_height

            # text_bbox = font.getbbox(active_subtitle)
            # text_width, text_height = text_bbox[2] - text_bbox[0], text_bbox[3] - text_bbox[1]
            # pos_x = (width - text_width) // 2
            # pos_y = height - text_height - 40

            # 描画の開始Y座標を計算 (下から40pxの余白)
            start_y = height - total_text_height - 40

            try:
                # 最も長い行を基準にROIを仮設定
                longest_line = max(wrapped_lines, key=lambda line: font.getbbox(line)[2])
                text_bbox = font.getbbox(longest_line)
                roi_w = text_bbox[2] - text_bbox[0]
                roi_x = (width - roi_w) // 2

                # bg_roi = frame[pos_y:pos_y+text_height, pos_x:pos_x+text_width]
                bg_roi = frame[start_y:start_y + total_text_height, roi_x:roi_x + roi_w]
                avg_bgr_color = cv2.mean(bg_roi)[:3]
                bg_luminance = get_luminance(avg_bgr_color)

                # --- ▼▼▼ 新しい色決定ロジック ▼▼▼ ---
                # 背景の明るさが中間(0.5)より暗いか明るいかでグループを決定
                if bg_luminance < 0.5:
                    # 背景が暗い -> 明るい色グループから最適色を探す
                    target_colors = LIGHT_COLORS
                    target_lums = LIGHT_LUMINANCES
                else:
                    # 背景が明るい -> 暗い色グループから最適色を探す
                    target_colors = DARK_COLORS
                    target_lums = DARK_LUMINANCES

                best_color = list(target_colors.values())[0] # グループの最初の色をデフォルトに
                max_contrast = 0
                for name, bgr_color in target_colors.items():
                    candidate_lum = target_lums[name]
                    contrast = get_contrast_ratio(bg_luminance, candidate_lum)
                    if contrast > max_contrast:
                        max_contrast = contrast
                        best_color = bgr_color

                text_color = best_color
                # --- ▲▲▲ ロジックここまで ▲▲▲ ---

            except cv2.error:
                text_color = (255, 255, 255)

            pil_img = Image.fromarray(cv2.cvtColor(frame, cv2.COLOR_BGR2RGB))
            draw = ImageDraw.Draw(pil_img)

            # draw.text((pos_x, pos_y), active_subtitle, font=font, fill=text_color[::-1])
            for i, line in enumerate(wrapped_lines):
                line_bbox = font.getbbox(line)
                line_width = line_bbox[2] - line_bbox[0]
                
                # 各行を中央揃えにする
                pos_x = (width - line_width) // 2
                pos_y = start_y + (i * line_height)
                
                draw.text((pos_x, pos_y), line, font=font, fill=text_color[::-1]) # text_colorはBGRなのでRGBに変換
                
            frame = cv2.cvtColor(np.array(pil_img), cv2.COLOR_RGB2BGR)

        output_video.write(frame)

    input_video.release()
    output_video.release()
    print(f"\n✅ 字幕の焼き付けが完了しました。")

    print("\n音声の結合処理を開始します...")
    final_video_path = os.path.join('project/output_video', f"final_{file_name}")
    !ffmpeg -i "{video_only_path}" -i "{input_video_path}" -c:v copy -c:a aac -map 0:v -map 1:a -y "{final_video_path}" -hide_banner -loglevel error
    os.remove(video_only_path)

    print(f"✅ 音声付きの最終動画 '{final_video_path}' を作成しました。")
    print("左のファイルブラウザからダウンロードしてください。")
```
